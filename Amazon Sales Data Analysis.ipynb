```python
#change dtypes
df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')


df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64')

df['discount_percentage'] = df['discount_percentage'] * 0.01

df['discount_percentage']
```

```python
sns.set_style(style='dark')
sns.set_palette(palette="crest")
```

```python
main_sub = NDF[['category_A', 'category_B', 'product_name']]

main_sub = main_sub.rename(columns={'category_A' :'Main Category', 'category_B' : 'Subcategory', 'product_name':'Product'})

main_sub_pivot = pd.pivot_table(main_sub, index=['Main Category', 'Subcategory'], aggfunc='count')

main_sub_pivot
```

```python
NDF['category_A'].value_counts().head(1).rename_axis('category_A').reset_index(name='counts')
```

```python
most_main_items = NDF['category_A'].value_counts().head(3).rename_axis('category_A').reset_index(name='counts')

most_sub_items = NDF['category_B'].value_counts().head(10).rename_axis('category_B').reset_index(name='counts')

fig, ax = plt.subplots(2, 1, figsize=(10,12))
plt.subplots_adjust(hspace = 0.3)
fig.suptitle('Most Amount of Products by Category', fontweight='heavy', size='x-large')

sns.barplot(ax=ax[0], data=most_main_items, x='counts', y='category_A')
ax[0].set_title('Most Products by Main Category', fontweight='bold')
ax[0].set_xlabel('Number of products sold', fontweight='bold')
ax[0].set_ylabel('Product Main Category', fontweight='bold')
ax[0].bar_label(ax[0].containers[0])

sns.barplot(ax=ax[1], data=most_sub_items, x='counts', y='category_B')
ax[1].set_title('Most Products by Subcategory', fontweight='bold')
ax[1].set_xlabel('Number of products sold', fontweight='bold')
ax[1].set_ylabel('Product Subcategory', fontweight='bold')
ax[1].bar_label(ax[1].containers[0])

plt.show()
```
